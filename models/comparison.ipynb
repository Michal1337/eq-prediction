```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def add_f1(df):
    df["f1"] = 2 * (df["precision"] * df["recall"]) / (df["precision"] + df["recall"])
    df["val_f1"] = 2 * (df["val_precision"] * df["val_recall"]) / (df["val_precision"] + df["val_recall"])
    return df
```

```python
df1 = pd.read_csv('history_model_v1.csv')
df2 = pd.read_csv('history_model_v2.csv')
df3 = pd.read_csv('history_model_v3.csv')
```

```python
df1 = add_f1(df1)
df2 = add_f1(df2)
df3 = add_f1(df3)
```

```python
df1.head()
```

```python
plt.plot(df1['loss'], label='Training loss model1')
plt.plot(df1['val_loss'], label='Validation loss model1')
plt.plot(df2['loss'], label='Training loss model2')
plt.plot(df2['val_loss'], label='Validation loss model2')
plt.plot(df3['loss'], label='Training loss model3')
plt.plot(df3['val_loss'], label='Validation loss model3')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(df1['f1'], label='Training F1 model1')
plt.plot(df1['val_f1'], label='Validation F1 model1')
plt.plot(df2['f1'], label='Training F1 model2')
plt.plot(df2['val_f1'], label='Validation F1 model2')
plt.plot(df3['f1'], label='Training F1 model3')
plt.plot(df3['val_f1'], label='Validation F1 model3')
plt.xlabel('Epoch')
plt.ylabel('F1')
plt.legend()
plt.show()
```
